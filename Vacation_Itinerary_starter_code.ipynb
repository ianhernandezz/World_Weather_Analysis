## Deliverable 3. Create a Travel Itinerary Map.

```python
# Dependencies and Setup
import pandas as pd
import requests
import gmaps
import numpy as np
import sys
sys.path.append("..")
# Import API key
from config import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
selected_cities = pd.read_csv("Vacation_Search/Vacation_Search.csv")
selected_cities
```

```python
# 1. Read the WeatherPy_vacation.csv into a DataFrame.
vacation_df = pd.read_csv("Vacation_Search/Vacation_Search.csv")
vacation_df.head()
```

```python
# 2. Using the template add the city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Weather Description</dt><dd>{Current Description}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# 3a. Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in vacation_df.iterrows()]

# 3b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = vacation_df[["Lat", "Lng"]]
```

```python
# 4a. Add a marker layer for each city to the map.
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(markers)
# 4b. Display the figure
fig
```

```python
# From the map above pick 4 cities and create a vacation itinerary route to travel between the four cities. 
# 5. Create DataFrames for each city by filtering the 'vacation_df' using the loc method. 
# Hint: The starting and ending city should be the same city.

vacation_start = vacation_df.loc[(vacation_df["Hotel Name"] == "Southern Sun Waterfront Cape Town")]
vacation_end = vacation_df.loc[(vacation_df["Hotel Name"] == "Southern Sun Waterfront Cape Town")]
vacation_stop1 = vacation_df.loc[(vacation_df["Hotel Name"] == "Oyster Bay House Rental")]
vacation_stop2 = vacation_df.loc[(vacation_df["Hotel Name"] == "Bredasdorp Country Manor")]
vacation_stop3 = vacation_df.loc[(vacation_df["Hotel Name"] == "Misty Waves Boutique Hotel")]
```

```python
# 6. Get the latitude-longitude pairs as tuples from each city DataFrame using the to_numpy function and list indexing.
#start = vacation_start[['Lat','Lng']].to_numpy(dtype=tuple)
#start_pair = [start[0,0], start[0,1]]
#end = vacation_end[['Lat','Lng']].to_numpy(dtype=tuple)
#end_pair = [end[0,0], end[0,1]]
#stop1 = vacation_stop1[["Lat","Lng"]].to_numpy(dtype=tuple)
#stop1_pair = [stop1[0,0], stop1[0,1]]
#stop2 = vacation_stop2[["Lat","Lng"]].to_numpy(dtype=tuple)
#stop2_pair = [stop2[0,0], stop2[0,1]]
#stop3 = vacation_stop3[["Lat","Lng"]].to_numpy(dtype=tuple)
#stop3_pair = [stop3[0,0], stop3[0,1]]
start = vacation_start['Lat'].values[0], vacation_start['Lng'].values[0]
end = vacation_end['Lat'].values[0], vacation_end['Lng'].values[0]
stop1 = vacation_stop1['Lat'].values[0], vacation_stop1['Lng'].values[0]
stop2 = vacation_stop2['Lat'].values[0], vacation_stop2['Lng'].values[0]
stop3 = vacation_stop3['Lat'].values[0], vacation_stop3['Lng'].values[0]

print(start)
print(end)
print(stop1)
print(stop2)
print(stop3)
```

```python
# 7. Create a direction layer map using the start and end latitude-longitude pairs,
# and stop1, stop2, and stop3 as the waypoints. The travel_mode should be "DRIVING", "BICYCLING", or "WALKING".

fig = gmaps.figure()
route = gmaps.directions_layer(
        start, end, waypoints=[stop1, stop2, stop3],
        travel_mode='DRIVING')
fig.add_layer(route)
fig
```

```python
# 8. To create a marker layer map between the four cities.
#  Combine the four city DataFrames into one DataFrame using the concat() function.
itinerary_df = pd.concat([vacation_start, vacation_stop1, vacation_stop2, vacation_stop3],ignore_index=True)
itinerary_df
```

```python
# 9 Using the template add city name, the country code, the weather description and maximum temperature for the city. 
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Weather Description</dt><dd>{Current Description}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# 10a Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in itinerary_df.iterrows()]

# 10b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = itinerary_df[["Lat", "Lng"]]
```

```python
# 11a. Add a marker layer for each city to the map.
fig = gmaps.figure()
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(markers)
# 11b. Display the figure
fig
```
